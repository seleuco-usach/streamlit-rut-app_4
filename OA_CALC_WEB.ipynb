```python
from io import BytesIO
import requests
import pandas as pd
import numpy as np



result = requests.get("https://mifuturo.cl/wp-content/uploads/2025/06/Oferta_Academica_2010_al_2025_SIES_02_06_2025_WEB_E.zip")
oa_histo = pd.read_csv(BytesIO(result.content),compression='zip', 
                 header=0, sep=';', 
                 quotechar='"', 
                 encoding='latin-1')

result = requests.get("https://mifuturo.cl/wp-content/uploads/2025/07/Matricula_2007_2025_WEB_15_07_2025.zip")
mat_histo = pd.read_csv(BytesIO(result.content),compression='zip', 
                 header=0, sep=';', 
                 quotechar='"', 
                 encoding='latin-1')
```

```python

oa_histo['anho'] = oa_histo['Año'].str.replace("OFE_", "", regex=False)

oa_histo['Nivel Global'] = np.where(oa_histo['Nivel Global'] == "Posgrado", 'Postgrado', 
         oa_histo['Nivel Global'])


oa_histo.columns
(
oa_histo.groupby(['Nivel Global','Código Único', 'anho'])
.size()
#.reset_index(name='counts')
.unstack()
.reset_index()
)


mat_histo['anho'] = mat_histo['AÑO'].str.replace("MAT_", "", regex=False)

mat_histo['anho'] = mat_histo['anho'].astype(int)

mat_histo['NOMBRE INSTITUCIÓN']

mat_tabla = (
mat_histo[mat_histo['anho']>2020]
.groupby(['NIVEL GLOBAL',
          'NOMBRE INSTITUCIÓN',
          'NOMBRE CARRERA', 
          'anho'])['TOTAL MATRÍCULA']
.sum()
#.reset_index(name='counts')
#.unstack()
.reset_index()
#.to_clipboard()
)
oa_histo.columns
(
oa_histo[oa_histo['Año']=="OFE_2025"]
.groupby(['Código Único','Área del conocimiento', 'Cine-F 13 Área'])
.size()
.reset_index(name="tot")
.to_clipboard()
)
#oa_histo.columns
```
